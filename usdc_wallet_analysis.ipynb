```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
from datetime import datetime
```

```python
ETHPLORER_API_KEY = "ENTER YOUR ETHPLORER API KEY"
ETHERSCAN_API_KEY = "ENTER YOUR ETHERSCAN API KEY"
USDC_CONTRACT = "0xa0b86991c6218b36c1d19d4a2e9eb0ce3606eb48"
TOP_N_WALLETS = 10
```

```python
# Known DeFi protocol addresses (add more as needed)
DEFI_PROTOCOLS = {
    "UniswapV2 Router": "0x7a250d5630b4cf539739df2c5dacf5a4438f3b23",
    "Aave Lending Pool": "0x7d2768dE32b0b80b7a3454c06BdAcA8B0e7bC6e3",
    "Compound cUSDC": "0x39aa39c021dfbae8fac545936693ac917d5e7563",
    # Add more protocol addresses as needed
}
```

```python
def get_top_usdc_holders():
    url = f"https://api.ethplorer.io/getTopTokenHolders/{USDC_CONTRACT}?apiKey={ETHPLORER_API_KEY}&limit={TOP_N_WALLETS}"
    resp = requests.get(url)
    data = resp.json()
    holders = [h['address'] for h in data.get('holders', [])]
    return holders

def get_wallet_usdc_transfers_etherscan(address):
    url = (
        f"https://api.etherscan.io/api"
        f"?module=account"
        f"&action=tokentx"
        f"&contractaddress={USDC_CONTRACT}"
        f"&address={address}"
        f"&startblock=0"
        f"&endblock=99999999"
        f"&sort=asc"
        f"&apikey={ETHERSCAN_API_KEY}"
    )
    resp = requests.get(url)
    data = resp.json()
    if data.get('status') == "1":
        return pd.DataFrame(data['result'])
    else:
        return pd.DataFrame()

def get_wallet_ethplorer_stats(address):
    url = f"https://api.ethplorer.io/getAddressInfo/{address}?apiKey={ETHPLORER_API_KEY}"
    resp = requests.get(url)
    return resp.json()

def analyze_wallet(address):
    print(f"Analyzing wallet: {address}")
    # Get USDC transfer data from Etherscan
    df_usdc = get_wallet_usdc_transfers_etherscan(address)
    time.sleep(0.25)  # To avoid rate limiting

    if not df_usdc.empty:
        df_usdc['timestamp'] = pd.to_datetime(df_usdc['timeStamp'], unit='s')
        df_usdc['direction'] = df_usdc.apply(lambda row: 'in' if row['to'].lower() == address.lower() else 'out', axis=1)
        df_usdc['value'] = df_usdc['value'].astype(float) / (10 ** 6)  # USDC has 6 decimals

        # Basic stats
        freq = df_usdc.groupby('direction').size().to_dict()
        volume = df_usdc.groupby('direction')['value'].sum().to_dict()
        first = df_usdc['timestamp'].min()
        last = df_usdc['timestamp'].max()
        active_days = (last - first).days if not pd.isnull(last) else 0

        protocol_touches = []
        for proto, proto_addr in DEFI_PROTOCOLS.items():
            if any((df_usdc['to'].str.lower() == proto_addr.lower()) | (df_usdc['from'].str.lower() == proto_addr.lower())):
                protocol_touches.append(proto)
    else:
        freq = volume = {}
        protocol_touches = []
        active_days = 0

    # Optionally, get extra stats from Ethplorer (such as ETH balance, token balances)
    ethplorer_stats = get_wallet_ethplorer_stats(address)
    eth_balance = ethplorer_stats.get('ETH', {}).get('balance', None)

    return {
        "address": address,
        "usdc_freq": freq,
        "usdc_volume": volume,
        "active_days": active_days,
        "protocols": protocol_touches,
        "df_usdc": df_usdc,
        "eth_balance": eth_balance
    }

def plot_stats(results):
    df_stats = pd.DataFrame([{
        "address": r["address"][:10] + "...",
        "in_tx": r["usdc_freq"].get('in', 0),
        "out_tx": r["usdc_freq"].get('out', 0),
        "in_volume": r["usdc_volume"].get('in', 0.0),
        "out_volume": r["usdc_volume"].get('out', 0.0),
        "active_days": r["active_days"],
        "eth_balance": r["eth_balance"],
        "protocols": ', '.join(r["protocols"])
    } for r in results])

    print("\nSummary stats:")
    print(df_stats)

    if df_stats.empty or ((df_stats['in_tx'] == 0) & (df_stats['out_tx'] == 0)).all():
        print("No USDC transfer data found for these wallets. No plots to display.")
        return

    sns.set(style="whitegrid")
    plt.figure(figsize=(12, 6))
    df_stats_melted = df_stats.melt(id_vars=['address'], value_vars=['in_tx', 'out_tx'], 
                                   var_name='Transaction Type', value_name='Count')
    sns.barplot(data=df_stats_melted, x="address", y="Count", hue="Transaction Type")
    plt.title("USDC Transfer Counts per Wallet")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    plt.figure(figsize=(12, 6))
    df_stats_melted = df_stats.melt(id_vars=['address'], value_vars=['in_volume', 'out_volume'], 
                                   var_name='Direction', value_name='Volume (USDC)')
    sns.barplot(data=df_stats_melted, x="address", y="Volume (USDC)", hue="Direction")
    plt.title("USDC Transfer Volume per Wallet")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("\nDeFi protocol interactions:")
    print(df_stats[['address', 'protocols']])

def main():
    print("Fetching top USDC holders from Ethplorer...")
    top_wallets = get_top_usdc_holders()
    results = []
    for addr in top_wallets:
        stats = analyze_wallet(addr)
        results.append(stats)
    plot_stats(results)
    return results

if __name__ == "__main__":
    results = main()
```

```python
# Period: Beginning of 2025 to now
START_DATE = datetime(2025, 1, 1)
END_DATE = datetime.now()

def get_top_usdc_holders():
    """Fetch top USDC holders from Ethplorer"""
    url = f"https://api.ethplorer.io/getTopTokenHolders/{USDC_CONTRACT}?apiKey={ETHPLORER_API_KEY}&limit={TOP_N_WALLETS}"
    resp = requests.get(url)
    data = resp.json()
    holders = []
    for h in data.get('holders', []):
        holders.append({
            'address': h['address'],
            'current_balance': float(h['balance']) / (10 ** 6),  # Convert to USDC (6 decimals)
            'share': h.get('share', 0)
        })
    return holders

def get_wallet_usdc_transfers(address):
    """Get all USDC transfers for a wallet"""
    url = (
        f"https://api.etherscan.io/api"
        f"?module=account"
        f"&action=tokentx"
        f"&contractaddress={USDC_CONTRACT}"
        f"&address={address}"
        f"&startblock=0"
        f"&endblock=99999999"
        f"&sort=asc"
        f"&apikey={ETHERSCAN_API_KEY}"
    )
    resp = requests.get(url)
    data = resp.json()
    if data.get('status') == "1":
        return pd.DataFrame(data['result'])
    else:
        return pd.DataFrame()

def analyze_wallet_period_change(wallet_info):
    """Calculate overall change for a wallet in the specified period"""
    address = wallet_info['address']
    print(f"Analyzing {address[:10]}...")
    
    # Get transfers
    df_transfers = get_wallet_usdc_transfers(address)
    time.sleep(0.25)  # Rate limiting
    
    if df_transfers.empty:
        return {
            'address': address,
            'address_short': address[:6] + '...' + address[-4:],
            'current_balance': wallet_info['current_balance'],
            'net_flow_2025': 0,
            'balance_start_2025': wallet_info['current_balance'],
            'balance_change': 0
        }
    
    # Process transfers
    df_transfers['timestamp'] = pd.to_datetime(df_transfers['timeStamp'], unit='s')
    df_transfers['value'] = df_transfers['value'].astype(float) / (10 ** 6)  # USDC decimals
    
    # Filter to 2025 period
    df_2025 = df_transfers[
        (df_transfers['timestamp'] >= START_DATE) & 
        (df_transfers['timestamp'] <= END_DATE)
    ].copy()
    
    # Calculate net flow in 2025 (positive = more incoming than outgoing)
    if not df_2025.empty:
        df_2025['flow'] = df_2025.apply(
            lambda row: row['value'] if row['to'].lower() == address.lower() else -row['value'], 
            axis=1
        )
        net_flow_2025 = df_2025['flow'].sum()
    else:
        net_flow_2025 = 0
    
    # Calculate balance at start of 2025
    df_before_2025 = df_transfers[df_transfers['timestamp'] < START_DATE]
    if not df_before_2025.empty:
        df_before_2025['flow'] = df_before_2025.apply(
            lambda row: row['value'] if row['to'].lower() == address.lower() else -row['value'], 
            axis=1
        )
        balance_start_2025 = df_before_2025['flow'].sum()
    else:
        balance_start_2025 = 0
    
    # Current balance is what we got from the API
    current_balance = wallet_info['current_balance']
    
    # Balance change = current - start of 2025
    balance_change = current_balance - balance_start_2025
    
    return {
        'address': address,
        'address_short': address[:6] + '...' + address[-4:],
        'current_balance': current_balance,
        'net_flow_2025': net_flow_2025,
        'balance_start_2025': balance_start_2025,
        'balance_change': balance_change
    }

def create_summary_graphs(results_df):
    """Create two summary graphs: Net Flow and Balance Change"""
    
    # Set up the plot style
    plt.style.use('default')
    sns.set_palette("husl")
    
    # Create figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 8))
    fig.suptitle('Top 10 USDC Wallets - 2025 Summary Analysis', fontsize=16, fontweight='bold')
    
    # Sort by absolute values for better visualization
    results_df_sorted = results_df.reindex(results_df['net_flow_2025'].abs().sort_values(ascending=False).index)
    
    # Graph 1: Net Flow in 2025
    colors1 = ['green' if x >= 0 else 'red' for x in results_df_sorted['net_flow_2025']]
    bars1 = ax1.bar(results_df_sorted['address_short'], results_df_sorted['net_flow_2025'], 
                    color=colors1, alpha=0.7, edgecolor='black', linewidth=0.5)
    
    ax1.set_title('Net USDC Flow in 2025\n(Positive = Net Inflow, Negative = Net Outflow)', 
                  fontsize=14, fontweight='bold')
    ax1.set_xlabel('Wallet Address')
    ax1.set_ylabel('Net Flow (USDC)')
    ax1.tick_params(axis='x', rotation=45)
    ax1.grid(True, alpha=0.3, axis='y')
    ax1.axhline(y=0, color='black', linestyle='-', linewidth=1)
    
    # Add value labels on bars
    for bar, value in zip(bars1, results_df_sorted['net_flow_2025']):
        height = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2., height,
                f'{value:,.0f}', ha='center', va='bottom' if height >= 0 else 'top',
                fontsize=10, fontweight='bold')
    
    # Graph 2: Overall Balance Change
    results_df_sorted2 = results_df.reindex(results_df['balance_change'].abs().sort_values(ascending=False).index)
    colors2 = ['blue' if x >= 0 else 'orange' for x in results_df_sorted2['balance_change']]
    bars2 = ax2.bar(results_df_sorted2['address_short'], results_df_sorted2['balance_change'], 
                    color=colors2, alpha=0.7, edgecolor='black', linewidth=0.5)
    
    ax2.set_title('Overall Balance Change (2025)\n(Current Balance - Start of 2025)', 
                  fontsize=14, fontweight='bold')
    ax2.set_xlabel('Wallet Address')
    ax2.set_ylabel('Balance Change (USDC)')
    ax2.tick_params(axis='x', rotation=45)
    ax2.grid(True, alpha=0.3, axis='y')
    ax2.axhline(y=0, color='black', linestyle='-', linewidth=1)
    
    # Add value labels on bars
    for bar, value in zip(bars2, results_df_sorted2['balance_change']):
        height = bar.get_height()
        ax2.text(bar.get_x() + bar.get_width()/2., height,
                f'{value:,.0f}', ha='center', va='bottom' if height >= 0 else 'top',
                fontsize=10, fontweight='bold')
    
    plt.tight_layout()
    plt.show()
    
    # Print summary table
    print("\n" + "="*100)
    print("SUMMARY TABLE")
    print("="*100)
    print(f"Period: {START_DATE.strftime('%Y-%m-%d')} to {END_DATE.strftime('%Y-%m-%d')}")
    print()
    print(f"{'Wallet':<12} {'Current Balance':<15} {'Start 2025':<15} {'Net Flow 2025':<15} {'Balance Change':<15}")
    print("-" * 100)
    
    for _, row in results_df.iterrows():
        print(f"{row['address_short']:<12} {row['current_balance']:>13,.0f} {row['balance_start_2025']:>13,.0f} "
              f"{row['net_flow_2025']:>13,.0f} {row['balance_change']:>13,.0f}")
    
    # Overall statistics
    total_net_flow = results_df['net_flow_2025'].sum()
    total_balance_change = results_df['balance_change'].sum()
    
    print("-" * 100)
    print(f"{'TOTAL':<12} {'':<15} {'':<15} {total_net_flow:>13,.0f} {total_balance_change:>13,.0f}")
    print("\nKey Insights:")
    print(f"• Total net flow across top 10 wallets: {total_net_flow:,.0f} USDC")
    print(f"• Total balance change across top 10 wallets: {total_balance_change:,.0f} USDC")
    print(f"• Wallets with net inflow: {len(results_df[results_df['net_flow_2025'] > 0])}")
    print(f"• Wallets with net outflow: {len(results_df[results_df['net_flow_2025'] < 0])}")

def main():
    """Main function"""
    print("USDC Top 10 Wallets - 2025 Analysis")
    print("="*50)
    
    # Get top wallets
    print("Fetching top USDC holders...")
    top_wallets = get_top_usdc_holders()
    
    if not top_wallets:
        print("Failed to fetch wallets. Check API configuration.")
        return
    
    print(f"Analyzing {len(top_wallets)} wallets...")
    
    # Analyze each wallet
    results = []
    for wallet in top_wallets:
        result = analyze_wallet_period_change(wallet)
        results.append(result)
    
    # Convert to DataFrame
    results_df = pd.DataFrame(results)
    
    # Create graphs
    create_summary_graphs(results_df)
    
    # Save results
    output_file = f"usdc_summary_{START_DATE.strftime('%Y%m%d')}_{END_DATE.strftime('%Y%m%d')}.csv"
    results_df.to_csv(output_file, index=False)
    print(f"\nResults saved to: {output_file}")
    
    return results_df

if __name__ == "__main__":
    results = main()
```
